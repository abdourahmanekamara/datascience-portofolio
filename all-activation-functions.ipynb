# Toutes les principales fonctions d'activations, à savoir:
- Sigmoid
- Tangente hyperbolique(Tanh)
- Unité de rectification Linéaire(ReLU)
- LeakyReLU
- Unité linéaire exponentiel(ELU)

```python

import numpy as np
import pandas as pd
from neuralnet.activations import Sigmoid,\
                                    Tanh,\
                                    ReLU,\
                                    LeakyReLU,\
                                    ELU
```

# Predictions

```python
z = np.linspace(-10,10,100)
```

```python
sig = Sigmoid()
sig.predict(z)
t = Tanh()
t.predict(z)
r = ReLU()
r.predict(z)
l = LeakyReLU()
l.predict(z)
e = ELU()
e.predict(z)
pd.DataFrame(   np.c_[sig.out,t.out,
                       r.out,
                       l.out,
                       e.out],columns=[
                           'Sigmoid',
                           'Tanh',
                           'ReLU',
                           'LeakyReLU',
                           'ELU'],
             index=z).plot()
```

# Dérivées

```python

pd.DataFrame(   np.c_[sig.derivative,t.derivative,
                       r.derivative,
                       l.derivative,
                       e.derivative],columns=[
                           'Sigmoid',
                           'Tanh',
                           'ReLU',
                           'LeakyReLU',
                           'ELU'],
             index=z).plot()
```
